```python
import sqlite3
import pandas as pd

# Specify the full path to the SQLite database file
db_file_path = r'C:\Users\leors\Project-4-Credit-Worthiness\Resources\credit_worthiness.sqlite'

# File paths for the CSV files
test_csv_path = r'C:\Users\leors\Project-4-Credit-Worthiness\Resources\test_final.csv'
train_csv_path = r'C:\Users\leors\Project-4-Credit-Worthiness\Resources\train_final.csv'

# Create an SQLite database (or connect to an existing one) at the specified path
db_conn = sqlite3.connect(db_file_path)

# Create tables for test and train data (only two tables: TestTable and TrainTable)
create_test_table = """
CREATE TABLE IF NOT EXISTS TestTable (
    ID INTEGER PRIMARY KEY,
    Customer_ID TEXT,
    Month TEXT,
    Name TEXT,
    Age INTEGER,
    SSN TEXT,
    Occupation TEXT,
    Annual_Income REAL,
    Monthly_Inhand_Salary REAL,
    Num_Bank_Accounts INTEGER,
    Num_Credit_Card INTEGER,
    Interest_Rate REAL,
    Num_of_Loan INTEGER,
    Type_of_Loan TEXT,
    Delay_from_due_date INTEGER,
    Num_of_Delayed_Payment INTEGER,
    Changed_Credit_Limit REAL,
    Num_Credit_Inquiries INTEGER,
    Credit_Mix TEXT,
    Outstanding_Debt REAL,
    Credit_Utilization_Ratio REAL,
    Credit_History_Age REAL,
    Payment_of_Min_Amount REAL,
    Total_EMI_per_month REAL,
    Amount_invested_monthly REAL,
    Payment_Behaviour TEXT,
    Monthly_Balance REAL,
    Credit_Score INTEGER,  
    Occupation_Num INTEGER,  
    Credit_Mix_Num INTEGER,  
    Payment_of_Min_Amount_Num INTEGER,  
    Payment_Behaviour_Num INTEGER  
);
"""

create_train_table = """
CREATE TABLE IF NOT EXISTS TrainTable (
    ID INTEGER PRIMARY KEY,
    Customer_ID TEXT,
    Month TEXT,
    Name TEXT,
    Age INTEGER,
    SSN TEXT,
    Occupation TEXT,
    Annual_Income REAL,
    Monthly_Inhand_Salary REAL,
    Num_Bank_Accounts INTEGER,
    Num_Credit_Card INTEGER,
    Interest_Rate REAL,
    Num_of_Loan INTEGER,
    Type_of_Loan TEXT,
    Delay_from_due_date INTEGER,
    Num_of_Delayed_Payment INTEGER,
    Changed_Credit_Limit REAL,
    Num_Credit_Inquiries INTEGER,
    Credit_Mix TEXT,
    Outstanding_Debt REAL,
    Credit_Utilization_Ratio REAL,
    Credit_History_Age REAL,
    Payment_of_Min_Amount REAL,
    Total_EMI_per_month REAL,
    Amount_invested_monthly REAL,
    Payment_Behaviour TEXT,
    Monthly_Balance REAL,
    Credit_Score INTEGER,  
    Occupation_Num INTEGER,  
    Credit_Mix_Num INTEGER,  
    Payment_of_Min_Amount_Num INTEGER,  
    Payment_Behaviour_Num INTEGER  
);
"""

# Execute the SQL commands to create the tables
db_conn.execute(create_test_table)
db_conn.execute(create_train_table)

# Read data from CSV files into DataFrames
test_data = pd.read_csv(test_csv_path)
train_data = pd.read_csv(train_csv_path)

# Insert data into the SQLite database
test_data.to_sql('TestTable', db_conn, if_exists='replace', index=False)
train_data.to_sql('TrainTable', db_conn, if_exists='replace', index=False)

# Commit changes and close the database connection
db_conn.commit()
db_conn.close()
```
